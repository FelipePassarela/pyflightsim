```python
position = glm.vec3(0, 0, 1)
orientation = glm.vec3(90, 0, 0)
scale = glm.vec3(1, 1, 1)

model = glm.mat4(1)
model = glm.translate(model, position)
model = glm.rotate(model, glm.radians(orientation.x), (1, 0, 0))
model = glm.rotate(model, glm.radians(orientation.y), (0, 1, 0))
model = glm.rotate(model, glm.radians(orientation.z), (0, 0, 1))
model = glm.scale(model, scale)

p = glm.vec4(1, 2, 5, 1)
p_world = model * p

print(model)
print(p_world)
```

```python
forward = glm.vec3(model[2])
eye = glm.vec3(model[3])
target = eye + forward
up = glm.vec3(model[1])
view = glm.lookAt(eye, target , up)

print(view)
print(view * p_world)
```

```python
VIEWPORT_W = 800
VIEWPORT_H = 600
projection = glm.perspective(glm.radians(60), VIEWPORT_W / VIEWPORT_H, 0.1, 100)
print(projection)
```

```python
mvp = projection * view * model
print(mvp)
```

```python
clip = mvp * p
clip
```

```python
visible = True

if clip.w < 0:  # behind camera
    visible = False
if not (-clip.w <= clip.x <= clip.w and -clip.w <= clip.y <= clip.w and -clip.w <= clip.z <= clip.w):
    visible = False

if visible and clip.w != 0:
    ndc = clip.xy / clip.w
    screen = glm.vec2(VIEWPORT_W / 2, VIEWPORT_H / 2) * (ndc + 1)
    print(ndc)
    print(screen)
```
